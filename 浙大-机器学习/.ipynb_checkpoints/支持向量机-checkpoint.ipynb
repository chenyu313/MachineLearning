# 线性可分定义

## 线性可分 - Linear Separable

&emsp;&emsp;存在一条直线可将目标区分。如果一个数据集是线性可分的，则存在无穷个超平面将各个类别分开。


## 线性不可分 - Nonlinear Separable

&emsp;&emsp;不存在一条直线将目标区分


线性可分定义：一个训练样本集{（xi,yi）}，i=1~N线性可分，是指存在（w，b）使得：  
- （1）当yi=+1时，w.T*wi + b>0
- （2）当yi=-1时，w.T*wi + b<0

# 问题描述

## 支持向量机算法

- 一、解决线性可分问题；
- 二、再将线性可分问题中获得的结论推广到线性不可分情况。

## 支持向量机寻找的最优分类直线应该满足

- （1）该直线分开了两类；
- （2）该直线最大化间隔（margin）；
- （3）该直线处于间隔的中间，到所有支持向量距离相等；


***
# 优化问题

如何用严格的数学 ==>寻找最优分类超平面过程 ==>写成一个最优化问题；  
假定训练样本集是线性可分的：支持向量机需要寻找的是最大化间隔（Margin）的超平面；


已知：训练样本集{（Xi,Yi）},i=1~N;  
待求：（w,b）

    上述问题是凸优化的二次规划；
    二次规划的定义：
- （1）目标函数是二次项；
- （2）限制条件是一次项。

凸优化问题一般当做一个解决了的问题，因为它只有唯一一个全局极值。

结论：一个优化问题是凸的，总能找到高效快速算法去解决它。

***
# 线性不可分的情况

以上的优化问题是没有解的，即不存在w和b满足上面所有N个限制条件；  
对于以上情况：放松限制条件的基本思路 ==>对每个训练样本及标签（Xi,Yi） ==>松弛变量

# 低维到高维的映射

&emsp;&emsp;支持向量机在扩大可选函数范围方面可谓独树一帜；  
支持向量机不直接产生函数，而是将特征空间由低维映射到高维，然后再高维的特征空间中用线性超平面对数据进行分类；


结论：将训练样本从低维映射到高维，可以增大线性可分的概率；
